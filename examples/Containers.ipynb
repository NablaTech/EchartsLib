# Containers

EchartsLib supports two types of plot containers, which could help you control multiple plots in one code cell.

## Figure

Figure is the basic container for all plots in echartslib.

There are two kinds of usage for Figure object.
- Directly call the plot object, and we will automatically put the plot to a default Figure object.
- Add plots object to Figure object


```python
# Directly call the plot object
import echartslib as elb
from sklearn.datasets import make_circles
pos,label = make_circles(n_samples=100, shuffle=True, noise=None, random_state=None, factor=0.8)
x,y = pos[:,0], pos[:,1]

plot_ = elb.scatterplot(x=x, y=y, label=label)
plot_
```

```python
# Put the plot object to figure, we could get the same result.
from echartslib import scatterplot

fig = elb.Figure()
plot_ = scatterplot(x=x, y=y, label=label)
fig.add(plot_)
fig
```

```python
# You also could add multiple plots in one figure.
from echartslib import lineplot, scatterplot

fig = elb.Figure()

scatter = scatterplot(x=x, y=y)
fig.add(scatter)

line = lineplot([-1,1], [-1,1])
fig.add(line)

fig
```

## Subplots

Subplots object supports user control the grid for each subplots and show them in single code cell.

```python
from echartslib import SubPlots, lineplot, scatterplot

scatter = scatterplot(x=x, y=y)
line = lineplot([-1,1], [-1,1])

sub = SubPlots([scatter, line], subplots_grid=[10,1], subplots_position=[[[0,5],[0,1]],[[6,10],[0,1]]])
sub
```
